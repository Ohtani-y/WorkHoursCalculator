```python
# pickleファイルからデータフレームを読み込むコード（コメントアウト）
df_hours_loaded = pd.read_pickle('data/月次合計.pkl')
df_absent_loaded = pd.read_pickle('data/欠勤履歴.pkl')
df_wk_loaded = pd.read_pickle('data/プロジェクト.pkl')
df_reg_loaded = pd.read_pickle('data/安衛法.pkl')
df_holiday_loaded = pd.read_pickle('data/有休取得実績.pkl')
df_vacation_loaded = pd.read_pickle('data/休暇取得.pkl')

display(df_hours_loaded.head())
display(df_absent_loaded.head())
display(df_wk_loaded.head())
display(df_reg_loaded.head())
display(df_holiday_loaded.head())
display(df_vacation_loaded.head())
```

```python
df_wk_loaded=df_wk_loaded[['日付', '就労所属名称', '社員番号', '社員名称', 'プロジェクト名称', '工数(ｈ)']]
```

```python
import pandas as pd

# 対象とする社員番号のリスト
target_employee_numbers = ['82043', '87002', '90060']

# 対象とする社員番号のデータを抽出
df_filtered = df_wk_loaded[df_wk_loaded['社員番号'].isin(target_employee_numbers)]

# 日付列をDatetime型に変換
df_filtered['日付'] = pd.to_datetime(df_filtered['日付'])

# 工数(ｈ)列を小数に変換
df_filtered['工数(ｈ)'] = df_filtered['工数(ｈ)'].astype(float)

# 日付列から年と月を抽出
df_filtered['年'] = df_filtered['日付'].dt.year
df_filtered['月'] = df_filtered['日付'].dt.month

# 月ごとのプロジェクトごとの工数を計算
project_monthly_summary = df_filtered.groupby(['社員番号', '社員名称', '就労所属名称', '年', '月', 'プロジェクト名称']).agg({'工数(ｈ)': 'sum'}).reset_index()

# 上位プロジェクトの数を指定
n_top_projects = 3  # 必要に応じてこの数を変更してください

# 各月ごとに上位プロジェクトを特定し、その割合を計算し、各列に整理する
def top_projects_percentage(group, n):
    total_hours = group['工数(ｈ)'].sum()
    top_projects = group.nlargest(n, '工数(ｈ)').reset_index(drop=True)
    result = {
        '社員番号': group['社員番号'].iloc[0],
        '社員名称': group['社員名称'].iloc[0],
        '就労所属名称': group['就労所属名称'].iloc[0],
        '年': group['年'].iloc[0],
        '月': group['月'].iloc[0],
        '総工数': total_hours
    }
    for i in range(n):
        result[f'プロジェクト{i+1}'] = top_projects.at[i, 'プロジェクト名称'] if len(top_projects) > i else None
        result[f'割合{i+1}'] = top_projects.at[i, '工数(ｈ)'] / total_hours * 100 if len(top_projects) > i else 0
    return pd.Series(result)

# 各月ごとに計算
result = project_monthly_summary.groupby(['社員番号', '社員名称', '就労所属名称', '年', '月']).apply(lambda x: top_projects_percentage(x, n_top_projects)).reset_index(drop=True)

# 結果を表示
display(result)
```
